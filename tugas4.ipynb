```python
import pandas as pd

# Path ke file CSV
file_path = 'Produksi Tanaman Buah-buahan, 2021-2023.csv'

# Membaca file CSV
data = pd.read_csv(file_path)

# Menampilkan data awal
data.head()
```

```python
# Menghapus kolom kosong dan baris tidak relevan
cleaned_data = data.dropna(how='all', axis=1).dropna(how='all', axis=0)
cleaned_data = cleaned_data.iloc[2:].reset_index(drop=True)

# Beri nama kolom sementara
cleaned_data.columns = ['Provinsi'] + [f'Column_{i}' for i in range(1, len(cleaned_data.columns))]

# Hapus baris dengan tanda '-' atau NaN
processed_data = cleaned_data[cleaned_data['Provinsi'].notnull()]
processed_data = processed_data.replace('-', None).dropna()
processed_data.head()
```

```python
from sklearn.model_selection import train_test_split

# Membagi data
train_data, temp_data = train_test_split(processed_data, test_size=0.4, random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

# Menampilkan ukuran data
print("Data Training:", train_data.shape)
print("Data Validasi:", val_data.shape)
print("Data Testing:", test_data.shape)
```

```python
# Simpan ke file CSV
train_data.to_csv('train_data.csv', index=False)
val_data.to_csv('val_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)
```

```python
# Buat file notebook
import nbformat as nbf

nb = nbf.v4.new_notebook()

# Tambahkan data ke notebook
nb.cells.append(nbf.v4.new_code_cell("# Data Training\ntrain_data"))
nb.cells.append(nbf.v4.new_code_cell("# Data Validasi\nval_data"))
nb.cells.append(nbf.v4.new_code_cell("# Data Testing\ntest_data"))

# Simpan notebook
with open('Data_Split_Dataset.ipynb', 'w') as f:
    nbf.write(nb, f)
```
